## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns

* Create a summary data frame to hold the results

```python
#Just as reference
purchase_data.head()
```

```python
#Retrieved the Item ID, Item Name, and Item Price columns
col_Item_ID = purchase_data['Item ID']
col_Item_Name = purchase_data['Item Name']
col_Item_Price = purchase_data ['Price']

#Converted columns to dataframe
popular_items_df = pd.DataFrame(col_Item_ID)
popular_items_df['Item Name'] = col_Item_Name
popular_items_df['Price'] = col_Item_Price

#dropping duplicates
popular_items_df_simplified =popular_items_df.drop_duplicates(subset = ['Item Name'])

#resetting indices
popular_items_df_simplified.reset_index(drop=True, inplace=True)
popular_items_df_simplified
```

```python
#Grouped by Item ID and Item Name
grouped_item_df = popular_items_df.groupby(["Item ID", "Item Name"])
```

```python
#purchase count of items dataframe
purchase_count_items = popular_items_df["Item Name"].value_counts() 
purchase_count_items_df = pd.DataFrame(purchase_count_items)
purchase_count_items_df
```

```python
purchase_count_items_df.reset_index(drop=True, inplace=True)
purchase_count_items_df.rename(columns = {'Item Name':'Purchase Count'}, inplace = True)
purchase_count_items_df
```

```python
#Column is extracted from dataframe
col_purchase_count_item = purchase_count_items_df["Purchase Count"]

#Column added to dataframe 'popular_items_df_simplified'
popular_items_df_simplified["Purchase Count"] = col_purchase_count_item
popular_items_df_simplified
```

```python
total_purchase_value_of_items = popular_items_df_simplified['Price'] * popular_items_df_simplified['Purchase Count']
total_purchase_value_of_items_df = pd.DataFrame(total_purchase_value_of_items, columns =['Total Purchase Value'])


total_purchase_value_of_items_df
```

```python
#Column is extracted from dataframe
col_total_purchase_value_of_items = total_purchase_value_of_items_df["Total Purchase Value"]

#Column added to dataframe 'popular_items_df_simplified'
popular_items_df_simplified["Total Purchase Value"] = col_total_purchase_value_of_items

popular_items_df_simplified
```
